````python
import pandas as pd
data=pd.read_csv("C:/Users/tarun/Downloads/Outlet.csv")

mean_mrp = data.groupby(['Outlet_Location_Type', 'Outlet_Size'])['Item_MRP'].mean().reset_index()

# Display the result
print(mean_mrp)
````

````python
mrp=data.groupby(['Outlet_Location_Type','Outlet_Size'])['Item_MRP'].mean().reset_index()
mrp
````



#calculate the min,max and median for Item_MRP and max, mean for Item_Visibility 
<br> ``` hint: use agg```

````python

# Assuming you have a dataframe called df
# Group by 'Outlet_Location_Type' and 'Outlet_Size' and then apply aggregation
agg_result = data.groupby(['Outlet_Location_Type', 'Outlet_Size']).agg(
    Item_MRP_min=('Item_MRP', 'min'),
    Item_MRP_max=('Item_MRP', 'max'),
    Item_MRP_median=('Item_MRP', 'median'),
    Item_Visibility_max=('Item_Visibility', 'max'),
    Item_Visibility_mean=('Item_Visibility', 'mean')
).reset_index()

# Display the result
agg_result
````
